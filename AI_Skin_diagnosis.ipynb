```python
import gradio as gr
from ultralytics import YOLO
from PIL import Image

# Load the trained YOLO model
model = YOLO('runs/detect/AI_skin_diagnosis2/weights/best.pt')  # Path to your trained model weights

# Function to make predictions
def diagnose_skin_condition(image):
    # Perform inference
    result = model(image)

    # Get bounding boxes and class labels
    output_image = result[5].plot()  # Draw detections on the image
    output = Image.fromarray(output_image)  # Convert to PIL Image for Gradio

    return output  # Return the annotated image with predictions

# Set up Gradio interface
gr_interface = gr.Interface(
    fn=diagnose_skin_condition,
    inputs=gr.Image(type="pil"),     # Accept image input
    outputs="image",                        # Display image output with bounding boxes
    title="AI Skin Diagnosis",
    description="Upload an image of a skin condition to get a diagnosis (e.g., acne, wrinkles, dark spots)."
)

# Launch the interface
gr_interface.launch()
```
